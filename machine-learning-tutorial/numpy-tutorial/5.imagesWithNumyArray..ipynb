```python
import numpy as np;
from skimage import io
from matplotlib import pyplot as plt
```

```python
# Lets create a numpy array from an image using scikit-image
myphoto = io.imread('jwm.jpg')
myphoto
```

```python
# Check for the type
type(myphoto)
```

```python
# Check for the sahpe
myphoto.shape
```

```python
# Using matplotlib to visualize the numpy array
plt.imshow(myphoto)
```

```python
# Lets retrieve just a portion of myphoto numpy array then visualize it with matplotlib
portionOfMyphoto = myphoto[0:500, 65:450]
plt.imshow(portionOfMyphoto)
```

```python
# keep Y and take a portion of X
portionOfMyphoto = myphoto[:, 65:450]
plt.imshow(portionOfMyphoto)
```

```python
# keep X and take a portion of Y
portionOfMyphoto = myphoto[0:500, :]
plt.imshow(portionOfMyphoto)
```

```python
# Lets reverse myphoto numpy array then visualize it with matplotlib
myphotoReversed = myphoto[::-1]
# myphotoReversed
plt.imshow(myphotoReversed)
```

```python
# Lets use numpy WHERE method to replace each value greater than 100 with 255 and the rest with 0
myphoto_masked = np.where(myphoto > 100, 255, 0)
plt.imshow(myphoto_masked)
```

```python
# Transpose
transposed_photo = myphoto[:, :, 0].T
plt.imshow(transposed_photo)
```
