### Correlogram Chart

```python
# imports
import matplotlib.pyplot as plt;
import seaborn as sns;
import numpy as np;
import pandas as pd;
```

```python
# imports data
iris= sns.load_dataset('iris')
iris.head()
```

```python
# Correlogram Chart using pairplot, a seaborn method. It will show you all the relationship between different data fields in your dataset
sns.pairplot(iris)
```

```python
# Adding diag_kind="kde" will show
sns.pairplot(iris, diag_kind="kde")
```

```python
# Adding linear regration to chart with this kind="reg"
sns.pairplot(iris, kind="reg")
```

```python
# Adding subgroup species to each
sns.pairplot(iris, hue="species")
```

```python
# Creating the metrix with selected columns
sns.pairplot(iris, vars=['sepal_length', 'sepal_width'], hue='species')
```
