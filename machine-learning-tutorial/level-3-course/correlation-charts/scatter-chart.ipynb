### Scatter Chart

```python
# imports
import matplotlib.pyplot as plt;
import seaborn as sns;
import numpy as np;
import pandas as pd;
```

```python
tips = sns.load_dataset('tips')
tips
```

1. Scarted Chart

```python
# using vanila matplotlib
fig, ax = plt.subplots(figsize=(7,5))
#ax.scatter(x="total_bill", y="tip", data=tips)
ax.scatter(x="total_bill", y="tip", data=tips, color="teal")
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Tip - Total Bill Scatter Chart")
plt.show()
```

```python
# Adding color based on sex
fig, ax = plt.subplots(figsize=(7,5))
colors = tips["sex"].map({"Male":"black", "Female": "red"})
ax.scatter(x="total_bill", y="tip", data=tips, color=colors)
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Tip - Total Bill Scatter Chart")
plt.show()
```

```python
# Adding color based on size
fig, ax = plt.subplots(figsize=(7,5))
colors = tips["size"].map({1:"black", 2: "red", 3: "green", 4: "yellow", 5: "red", 6: "blue"})
ax.scatter(x="total_bill", y="tip", data=tips, color=colors)
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Tip - Total Bill Scatter Chart")
plt.show()
```

```python
# adding c="tip" and cmap="Reds" will make the red color stronger based on how high is the tip
fig, ax = plt.subplots(figsize=(7,5))
ax.scatter(x="total_bill", y="tip", c="tip", cmap="Reds", data=tips)
ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Tip - Total Bill Scatter Chart")
plt.show()
```

```python
# Adding a colorbar
fig, ax = plt.subplots(figsize=(7,5))

im = ax.scatter(x="total_bill", y="tip", c="tip", cmap="Blues", data=tips)
fig.colorbar(im, ax=ax)

ax.set_xlabel("total_bill")
ax.set_ylabel("tip")
ax.set_title("Tip - Total Bill Scatter Chart")
plt.show()
```

```python
#using seaborn
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,10))
ax1.set_title("Tip - Total Bill Scatter Chart")
# ax2.set_title("Tip - Total Bill Scatter Chart")
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="sex", ax=ax1)
sns.scatterplot(data=tips, x="total_bill", y="tip", hue="size", ax=ax2)
plt.show()
```

```python
iris = sns.load_dataset("iris")
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10,10))
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", ax=ax1)
sns.scatterplot(data=iris, x="sepal_length", y="petal_length", hue="species", ax=ax2)
```
