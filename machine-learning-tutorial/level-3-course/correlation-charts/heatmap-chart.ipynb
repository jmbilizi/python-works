### HeatMap

It shows the relationship between two variables, one plotted on each axis using colors. With heatmap, you can observe how cell colors change across each axis  to identify patterns. The cells either contain color-coded categorical data or numerical data that is based on a color scale.

```python
# imports
import matplotlib.pyplot as plt;
import seaborn as sns;
import numpy as np;
import pandas as pd;
```

```python
iris = sns.load_dataset('iris')
iris
```

```python
# Using seaborn. The heatmap only works with numerical values
fig, ax = plt.subplots(figsize=(6, 6))
ax.set_title('Iris Heatmap Correlation Chart')
sns.heatmap(iris.drop(['species'], axis=1).corr(), annot=True, fmt=".2f")
```

```python
# Load flights dataset from sns
flights = sns.load_dataset('flights')
flights.head()
```

```python
# using pandas's Pivot function
new_flights = flights.pivot(index='month', columns='year', values='passengers')
new_flights
```

```python
# Pivot heatmap
fig, ax = plt.subplots(figsize=(10, 7))
sns.heatmap(new_flights, ax=ax)
```

```python
# Pivot heatmap with linewidths between.
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_title('Flight Reservation Chart')
sns.heatmap(new_flights, linewidths=1, ax=ax)
plt.show()
```

```python
# Pivot heatmap with linewidths between and annotation
fig, ax = plt.subplots(figsize=(10, 7))
ax.set_title('Flight Reservation Chart')
sns.heatmap(new_flights, annot=True, fmt='d', linewidths=1, ax=ax)
plt.show()
```
