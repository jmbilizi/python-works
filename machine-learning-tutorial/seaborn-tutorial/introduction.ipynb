```python
# Importing seaborn
import seaborn as sns;
```

```python
# Get list names of dummy data you auto import from seaborn
sns.get_dataset_names()
```

```python
# Using seaborn to import dammy data (But You can use or import any data you want)
taxisData = sns.load_dataset('taxis')
# type(data)
# data.shape
taxisData
```

```python
# To show correlation between data fields in pandas
taxisData.corr()
```

```python
# Chart with correlation between data fields in seaborn. annot (show numbers or no), cmap('YlGnBu' or 'icefire', etc...)
sns.heatmap(taxisData.corr(), annot=True, cmap="coolwarm")
```

```python
# Lets use seaborn to display chart (https://seaborn.pydata.org/examples/index.html)
sns.scatterplot(x="passengers", y="total", data=taxisData, hue="pickup_borough", size="distance", palette="YlGnBu")
```

```python
sns.histplot(taxisData["total"], kde=True, bins=15)
```

```python
# barplot or boxplot chart
sns.boxplot(data=taxisData, x="pickup_borough", y="total", palette="YlGnBu")
```

```python
# barplot chart
sns.barplot(data=taxisData, x="pickup_borough", y="total", palette="YlGnBu")
```

```python
# jointplot --> is a regretional chart that shows you where the concentration of values are.
sns.jointplot(data=taxisData, x="total", y="distance", kind="kde", fill=True)
```

```python
# pairplot
sns.pairplot(taxisData.select_dtypes(["number"]), hue="total")
```

```python
# clustering dataset chart  --> accept numeric only dataset (Great to look at when performing unsupervised learning)
numericOnlyTaxisData = taxisData.drop(['pickup', 'dropoff', 'color', 'payment', 'pickup_zone', 'dropoff_zone', 'pickup_borough', 'dropoff_borough'], axis=1)
sns.clustermap(numericOnlyTaxisData)
```
