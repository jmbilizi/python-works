## Sorting and Ranking DataFrame 

```python
# Here is how to import pandas in your project
import pandas as pd;
```

```python
df = pd.read_csv('../../titanic.csv', header=0)
```

1. Sorting by a column

```python
# Sort by ascending order of Age column values using sort_values method with ascenging flag
data = df
data.sort_values('Age', ascending=True).head(10)
```

```python
# Sort, filter top 10 rows then filter selected columns all at once with  methods chaining
data = df
data.sort_values('Age', ascending=True).head(10)[['Name','Age', 'Fare']]
```

```python
# Multi Sorting using sort_values
data = df
data.sort_values(['Sex', 'Age'], ascending=[True, False])[['Name', 'Sex', 'Age']]
```

```python
# Drop columns with missing values then sort data with method chaining
data = df
data.dropna().sort_values(['Sex', 'Age'], ascending=[True, False])[['Name', 'Sex', 'Age']]
```

```python
# Top 10 or n row with largest value for a given column 
data = df
data.nlargest(10, 'Age')[['Name', 'Age', 'Sex']]
```

```python
# Top 10 or n row with smallest value for a given column 
data = df
data.nsmallest(10, 'Age')[['Name', 'Age', 'Sex']]
```
