## Filtering DataFrame

```python
# Here is how to import pandas in your project
import pandas as pd;
```

```python
df = pd.read_csv('../../titanic.csv', header=0)
```

1. Using Just Boolean Selection

Consiste of creating a copy of a given colomn with a value of true or false in each row. Then use those values to filter.

```python
# Example of bolean selection
data = df
data['Sex'] == 'female'
```

```python
# Filtering table based of the bolean selection condition
data = df
femaleSex = data['Sex'] == 'female'
data[femaleSex]
```

```python
# Filter all passengers below the age of 16
data = df
data[data['Age'] > 60]
```

```python
# Multi filter. Filter all passenger above the age of 60 AND that are female
data = df
data[(data['Age'] > 60) & (data['Sex'] == 'female')]
```

```python
# Multi filter. Filter all passenger above the age of 60 OR that are female
data = df
data[(data['Age'] > 60) | (data['Sex'] == 'female')]
```

2. Using The Loc Method

```python
# Multi filter rows and reduced columnsusing Loc method
data = df
condition = (data['Age'] > 75) | (data['Sex'] <= 'female') | (data['Age'] <= 20)
data.loc[condition, ['Name', 'Age', 'Sex', 'Fare']]
```
