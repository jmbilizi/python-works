- Remove Columns

Lets assume we would like to remove Ticket and Fare colums. We can use the drop method by providing it with the list of strings as columns to be removed, and axis with value 1 which is the axis value for a column.

```python
# Import pandas
import pandas as pd;
```

```python
# load data from csv file into DataFrame
df = pd.read_csv('../../titanic.csv')
```

```python
# Remove these two columns From DataFrame without changing the original DataFrame
data = df.drop(['Ticket', 'Fare'], axis=1)
data.head()
# if you run this original DataFrame df.head(), those two columns should still be there
```

```python
# Remove these two columns from the original DataFrame, set inplace=True
df.drop(['Ticket', 'Fare'], axis=1, inplace=True)
df.head()
```
