```python
# Here is how to import pandas in your project
import pandas as pd;
```

```python
# load data from csv file into DataFrame
df = pd.DataFrame({'name': ['Alice', 'Janvier', 'Naomie', 'Kate', 'Bob', 'Jeff', 'Jack'], 'age': [25, 30, 11, 25, 40, 25, None], 'Sex': ['female', 'male', 'female', 'female', 'male', 'male', 'male'], 'city': ['Rockville', 'Germantown', 'Denver', 'Rome', None, 'Harrisonburg', None]})
df.head()
```

1. Create Dummy data By maping a column into another one using map method (Like mapping a Categorique string values into numbers)

```python
# Map categorique data Sex to create anothe column femaleGender with 1 or 0 (female or male)
data = pd.DataFrame({'name': ['Alice', 'Janvier', 'Naomie', 'Kate', 'Bob', 'Jeff', 'Jack'], 'age': [25, 30, 11, 25, 40, 25, None], 'Sex': ['female', 'male', 'female', 'female', 'male', 'male', 'male'], 'city': ['Rockville', 'Germantown', 'Denver', 'Rome', None, 'Harrisonburg', None]})
data['femaleGender'] = data['Sex'].map({'female': 1, 'male': 1}) # Create new columns with the map
data.drop(['Sex'], axis=1, inplace=True) # Remove the original Sex column
data
```

2. Create Dummy data Using get_dummies method

get_dummies method will automatically create columns for each unique or categorique value of the column

```python
# get_dummies
data = pd.DataFrame({'name': ['Alice', 'Janvier', 'Naomie', 'Kate', 'Bob', 'Jeff', 'Jack'], 'age': [25, 30, 11, 25, 40, 25, None], 'Sex': ['female', 'male', 'female', 'female', 'male', 'male', 'male'], 'city': ['Rockville', 'Germantown', 'Denver', 'Rome', None, 'Harrisonburg', None]})
pd.get_dummies(data, columns=['Sex'])
```

```python
# With get_dummies method for K possible values --> K-1 variables so we can eliminate one created column by adding drop_first=True
data = pd.DataFrame({'name': ['Alice', 'Janvier', 'Naomie', 'Kate', 'Bob', 'Jeff', 'Jack'], 'age': [25, 30, 11, 25, 40, 25, None], 'Sex': ['female', 'male', 'female', 'female', 'male', 'male', 'male'], 'city': ['Rockville', 'Germantown', 'Denver', 'Rome', None, 'Harrisonburg', None]})
pd.get_dummies(data, columns=['Sex'], drop_first=True)
```

```python
# To store the data in a different variable
new_data = pd.get_dummies(data, columns=['Sex'], drop_first=True)
new_data
```
