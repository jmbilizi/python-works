## Vector

Vector is a 1 dimensional tensor (1D)

### Using Vector in pytorch

```python
# Importing torch
import torch;
```

```python
# Vector operations
tensor_one = torch.tensor([2, 6, 12])
tensor_two = torch.tensor([1, 2, 3])

print(tensor_one + tensor_two) # Addition
print(tensor_one - tensor_two) # Substraction
print(tensor_one * tensor_two) # Multiplication
print(tensor_one - tensor_two) # Division

dot_product = torch.dot(tensor_one, tensor_two)
print(dot_product)
```

```python
# Vector operation with a Scalar (zero dimensional tensor)
tensor_1 = torch.tensor([2, 6, 12])

print(tensor_1 + 3) # 1. Addition
print(tensor_1 * 3) # Multiplication
print(tensor_1 - 1) # Substraction
print(tensor_1 / 2) # Division
```

```python
# Generating equally spaced value using linspace torch method
x = torch.linspace(0, 10, 100)
x
```

```python
# Using tensor with matplotlab
import matplotlib.pyplot as plt

x = torch.linspace(0, 10, 100)
y = torch.sin(x)

# plt.plot(x, y1, label='line 1')
plt.plot(x, y,)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My Simple Line Chart 1')
```

```python
# Using tensor with matplotlab 
import matplotlib.pyplot as plt

x = torch.linspace(0, 10, 100)
y = torch.exp(x)

plt.plot(x, y,)
plt.xlabel('x')
plt.ylabel('y')
plt.title('My Simple Line Chart 2')
```
